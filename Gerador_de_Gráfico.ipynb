```
import matplotlib.pyplot as plt
import numpy as np

# Dados da Tabela 1 do Relatório (Mensagem de 20 bits)
# Eixo X: Relação Sinal-Ruído (SNR) em dB
snr_db = np.array([0, -6, -12, -18, -21, -24, -27])
# Eixo Y: Número de Erros (NRZ)
erros_nrz = np.array([0, 1, 8, 14, 20, 20, 20])
# Eixo Y: Número de Erros (Manchester)
erros_manchester = np.array([0, 0, 2, 5, 9, 15, 20])
total_bits = 20

# 1. Cálculo da Taxa de Erro de Bit (BER)
ber_nrz = erros_nrz / total_bits
ber_manchester = erros_manchester / total_bits

# Configuração da figura com dois subplots (1 linha, 2 colunas)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle(f'Análise A3.1: Desempenho de Modulações sob Ruído (Mensagem de {total_bits} bits)', fontsize=16)

# --- GRÁFICO 1: Comparação de Erros (NRZ vs. Manchester) ---
# Título: GRÁFICO 1
ax1.plot(snr_db, erros_nrz, marker='o', linestyle='-', color='red', label='NRZ')
ax1.plot(snr_db, erros_manchester, marker='s', linestyle='--', color='blue', label='Manchester')

# Configurações do GRÁFICO 1
ax1.set_title('GRÁFICO 1: Número de Erros vs. SNR', fontsize=14)
ax1.set_xlabel('Relação Sinal-Ruído (SNR) em dB', fontsize=12)
ax1.set_ylabel(f'Número de Erros de Bit (Máx. {total_bits})', fontsize=12)
ax1.set_yticks(np.arange(0, total_bits + 1, 5))
ax1.grid(True, linestyle=':', alpha=0.6)
ax1.legend(loc='upper left')
ax1.set_xticks(snr_db)

# Marcadores para os pontos de falha total
ax1.axvline(x=-21, color='red', linestyle=':', linewidth=1, alpha=0.7)
ax1.axvline(x=-27, color='blue', linestyle=':', linewidth=1, alpha=0.7)


# --- GRÁFICO 2: Taxa de Erro de Bit (BER) vs. SNR (dB) ---
# Título: GRÁFICO 2
ax2.plot(snr_db, ber_nrz, marker='o', linestyle='-', color='red', label='NRZ')
ax2.plot(snr_db, ber_manchester, marker='s', linestyle='--', color='blue', label='Manchester')

# Configurações do GRÁFICO 2
ax2.set_title('GRÁFICO 2: Taxa de Erro de Bit (BER) vs. SNR (dB)', fontsize=14)
ax2.set_xlabel('Relação Sinal-Ruído (SNR) em dB', fontsize=12)
ax2.set_ylabel('Taxa de Erro de Bit (BER) [0 a 1]', fontsize=12)
ax2.set_yticks(np.arange(0, 1.1, 0.2)) # Eixo Y de 0 a 1, passos de 0.2
ax2.grid(True, linestyle=':', alpha=0.6)
ax2.legend(loc='upper left')
ax2.set_xticks(snr_db)


plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ajusta layout para a suptitle

# Descomente e execute a linha abaixo para salvar o gráfico como imagem:
# plt.savefig('graficos_comparacao_a31.png')

plt.show() # Mostrar os gráficos
```
